-4번 문제: 다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)

####s = '  first star   :   second star   :    third star  '
####l = ['first star', 'second star', 'third star']
####1) for ~ in 문을 이용한 코드를 작성하시오.


```python
## !) For ~ in 문 사용한 코드 
s = '  first star   :   second star   :    third star  '


l = s.split(':')
count=0

for a in l:
    l[count]=a.strip()
    count=count+1
    
print l
```

####2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '

l = s.split(':')

L=[a.strip() for a in l]
    
print L
```

##설명
1. 변수ㅣ에 s에 저장된 문자열을 ':' 로 split 하여 리스트 형태로 저장한다 그럼 공백이들어간 문자열들이 원소로 들어가있는데 이 원소들을 for 문을 이용하여 하나씩 꺼내와 strip() 함수를 이용하여 공백을 제거하고 count 변수를 추가하여 l[count] 로 인덱스를 바꿔가면서 공백이 제거된 문자열을 집어넣어준다.

2. 리스트 내포를 사용하여 for ~ in 을 사용하면 4줄인 코드를 L = [a.strip() for a in l] 한줄 코드로 변경하였다.

####1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
####>>> addall([1])

####>>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])

```python
## for ~ in 방식 
def addall(list):
    sum=0
    for a in list:
        sum=sum+a
        
    return sum

print addall([1])
print addall([1,2,3,4,5,6,7,8,9])
```

```python
## 리스트 내포  방식 
def addall(list):
    
    return sum([a for a in list ])    
   

print addall([1])
print addall([1,2,3,4,5,6,7,8,9])
```

####2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

####>>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])

```python
## for ~ in 방식
def addallodd(list):
    sum=0
    for a in list:
        if a % 2 == 1:
            sum = sum + a
    return sum

print addallodd([1])
print addallodd([1,2,3,4,5,6,7,8,9])
    
```

```python
## 리스트 내포 방식
def addallodd(list):
    
    return sum([a for a in list if a % 2])  

   
print addallodd([1])
print addallodd([1,2,3,4,5,6,7,8,9])
```

```python
# -*- coding:utf-8 -*-

import json        
import urllib2
import string
source = urllib2.urlopen("http://www.naver.com").read()  




##print source

    
def spword_remove(a):
    out=a
    spword=[]
    spword2=['�','nbsp','raquo']
             
    for l in string.punctuation:
        spword.append(l) 
        spword.extend(spword2)

      
    ##l=['�','nbsp','!','"','#','$','%','&',"'",'(',')','*','+',',','-','.','/',':',';','<','=','>','?','@','[','\\',']','^','_','`','{','|','}','~']
    
    for n in range(len(spword)):
        out=out.replace(spword[n],'')
        
    return out
   

        
def comment_remove(a):
    comment = False
    chkword =""
    out =""
    
    for s in a:
        chkword = chkword + s 
        
        
        if '<!--' in chkword:
            comment = True
            chkword=""
        elif '-->' in chkword:
            comment = False
            chkword=""
        elif comment == False:
            out = out + s
    return out


def comment_remove2(a):
    comment = False
    chkword =""
    out =""
    
    for s in a:
        chkword = chkword + s 
        
        
        if '<![CDATA' in chkword:
            comment = True
            chkword=""
        elif ']]>' in chkword:
            comment = False
            chkword=""
        elif comment == False:
            out = out + s
    return out

def style_remove(a):
    style = False
    chkword =""
    out =""
    
    for s in a:
        chkword = chkword + s 
        
        
        if '<style' in chkword:
            style = True
            chkword=""
        elif '/style>' in chkword:
            style = False
            chkword=""
        elif style == False:
            out = out + s
    return out

def extract_word(a):
    tag = False
    out = ""
    b = comment_remove2(a)
    c = comment_remove(b)
    d = style_remove(c)
    for s in d:
   
            if s == '<' :       ##읽은 문자열이 < 일경우 태그 로판단
                tag = True
            elif s == '>' :     ##읽은 문자가 > 일경우 태그가 끝난걸로 판단.
                tag = False    
            elif tag==False:  ## 위조건문을 지나고 만약 tag가 False면은 순수한 단어이기떄문에 저장한다.
                out = out + s

    return out



a=extract_word(source)        ## 추출된 단어들을 모은 문자열을 a변수에 저장하고
a=spword_remove(a)            ## 특수문자들을 제거한다.


l=a.split()                   ## 공백을 중심으로 잘라 저장한다


count = 0
for a in l:
    count = count +1
    print "단어 %d: %s   "  % (count,a) 
print "단어 개수 : " ,len(l)

d = {}
for i in l:
    if i not in d: 
        d[i]=1
    elif i in d:
        d[i]=d[i]+1


"""
## unicode(a , encoding = 'utf-8')
for i in l:
    if i not in d:
        key=unicode(i , encoding = 'utf-8')
        d[key]=1
    elif i in d:
        key=unicode(i , encoding = 'utf-8')
        d[key]=d[key]+1
    
print d,keys()
"""



print
print json.dumps(d,ensure_ascii=False)
```
