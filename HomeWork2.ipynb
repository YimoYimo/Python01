####2번 : 키보드로 정수값을 입력 받고 그 값이 양수인지 음수인지를 비트 연산자를 이용하여 판단하는 프로그램을 작성하시오.
####[참고] 가장 왼쪽에 있는 비트가 0이면 양수이며, 1이면 음수이다.
####[참고] 수업보충자료 (https://github.com/bluebibi/python-e-learning/blob/master/supplement.ipynb) 를 참고하시오.

```python
i=input("정수를 입력하세요")

num = len(bin(i))-2

if (i >> num) & 1 == True:
    print "음수입니다"
elif (i >> num) & 1 == False:
    print "양수입니다"
```

#### 설명: 정수를 입력하고  입력한 정수를 bin()함수를 사용해 2진수 문자열의 길이를 num 변수에 넣습니다(-2를 한이유는 0b때문).  입력한 정수를 num 만큼 비트이동 시키고 만약 끝까지 이동했는데도 1이 나온다면 조건문에 의해 true 이고 즉 음수로 판단합니다. 아닐경우 양수로 판단합니다.

####3번 : 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
####각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
####반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
####다음과 같은 실행 및 출력 결과가 도출되어야 한다.
####div_path('/usr/local/bin/python')
####['/usr/local/bin', 'python']
####div_path('/home/chulsoo/test.txt')
####['/home/chulsoo', 'test.txt']


```python
def div_path(s):
    l=s.split('/')
    name = l.pop(len(l)-1)
    root = '/'.join(l) 
    del l[:]
    l.append(root)
    l.append(name)
    print l
    
div_path('/usr/local/bin/python')    
div_path('/home/chulsoo/test.txt')
         
```

####설명 : 문자열을 split 함수를 사용하여 '/' 기준으로 잘라 l 리스트로 저장한다.  저장된 마지막 원소가 파일명이므로 pop 함수를 사용하여 name 변수에 저장하고 나머지 원소들을 다시 '/' 로 join 시켜서 root 변수에 저장한다.   처음 사용했던 l 리스트 원소들을 모두 제거한후 다시 append 함수를 사용하여 root 와, name 을 집어넣어 출력시킨다.

####4번. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
####인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
####함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
####반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
####반환하는 리스트는 정렬이 되어 있어야 한다.
####다음과 같은 실행 및 출력 결과가 도출되어야 한다.
####list_union([1, 2, 3], [1, 2, 4])
####        ->  [1, 2, 3, 4]
####list_union([-10, -5, 0, -1], [100, 9, 0, 9])
####         -> [-10, -5, -1, 0, 9, 100]
####list_union([0, 1, 2], [0, 1, 2])
####          ->[0, 1, 2]

####[참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
####[참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb):
    
    l=lista
    count = 0 
    
    while count < len(listb):
        
        if listb[count] in lista:
            count=count+1
        else:
            l.append(listb[count])
            count=count+1
    l.sort()
    
    print l        
                


list_union([-1,2,-3],[1,9,3,5,-6])
```

####설명: 첫번쨰 인자로 받은 lista 리스트를 l 변수에 넣는다. 그리고 반복문 조건으로 두번째 인자의 길이만큼 루프를 돌게 하여 두번쨰 리스트의 인덱스0부터 끝까지 값을 가져와 리스트l에 저장된 값과 중복이 되는지 if else 문으로 확인후 집어 넣는다. 반복문이 끝난후 sort() 함수를 사용하여 정렬을하고 출력한다.

```python
def gcd(a, b):
    while (b != 0):
        temp = a % b
        a = b
        b = temp
    return abs(a)

def lcm(a, b):
    gcd_value = gcd(a, b)
    if (gcd_value == 0): return 0 
    return abs( (a * b) / gcd_value )

start = 1
for i in range(1,21):
    start = lcm(start,i)
print start
```

####[Incremental Project] 웹 URL로 지정된 웹페이지를 문자열로 가져와 
####1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 
####2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.


```python
import urllib2

source = urllib2.urlopen("http://cse.kut.ac.kr/").read()

##print source 

def extract_word(a):
    tag = False
    
    out = ""
    
    for s in a:
            if s == '<' :       ##읽은 문자열이 < 일경우 태그 로판단
                tag = True
            elif s == '>'   :     ##읽은 문자가 > 일경우 태그가 끝난걸로 판단.
                tag = False    
            elif tag==False:  ## 위조건문을 지나고 만약 tag가 False면은 순수한 단어이기떄문에 저장한다.
                out = out + s

    return out

a=extract_word(source)        ## 추출된 단어들을 모은 문자열을 a변수에 저장하고

l=a.split()                   ## 공백을 중심으로 잘라 저장한다
count = 0
for a in l:
    count = count +1
    print "단어 %d: %s   "  % (count,a) 
print "단어 개수 : " ,len(l)


```

####설명 : extract_word 라는 함수안에 URL의 문자열을 받아온 source 변수를 인자로 넣고 '<' 를 읽을경우 tag 로 판단하여 out에 "" 를 저장 즉 빈공백을 저장하고  '>' 를읽어 태그가 아닌걸 판단할경우 문자열에서 읽은 문자 1개를 out에 저장시켜 단어들을 저장한다.
####split()함수를 사용하여 공백을 중심으로 잘라 l변수에 저장하고 for문을 돌려 출력합니다.
